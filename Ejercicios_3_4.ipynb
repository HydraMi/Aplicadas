```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 100, 1)
sin_n_div_n = np.sin(n) / n

plt.plot(n, sin_n_div_n, label=r'$\frac{\sin(n)}{n}$')
plt.hlines(0, 1, 100, colors='r', label=r'$y = 0$', linestyles='dashed')
plt.xlabel('n')
plt.ylabel(r'$\frac{\sin(n)}{n}$')
plt.title(r'Comportamiento de $\frac{\sin(n)}{n}$ para grandes valores de $n$')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

n_values = np.arange(1, 201)
a_n_values = (1 + 1 / n_values) ** n_values


plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n_values, label="a_n = (1 + 1/n)^n", color="blue")
plt.axhline(y=math.e, color='red', linestyle='--', label="y = e")
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("Plot de a_n vs n con la línea y = e")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

e = np.exp(1)
n_values = np.arange(1, 201)
a_n_values = [(1 + 1/n)**n for n in n_values]

for n, a_n in zip(n_values, a_n_values):
    if abs(a_n - e) < 0.01:
        print(f"El primer valor de n tal que |a_n - e| < 0.01 es: {n}")
        break
```

#$a).$#

```python
import numpy as np
import matplotlib.pyplot as plt

n_values = np.array([1, 2, 3, 4, 5])
a_n_values = n_values * np.sin(1 / n_values)

print("Valores de a_n para n = 1, 2, 3, 4, 5:")
for n, a_n in zip(n_values, a_n_values):
    print(f"a_{n} = {a_n:.6f}")
```

#$b).$#

```python
n_full = np.arange(1, 201)
a_n_full = n_full * np.sin(1 / n_full)
print("Valores de a_n para n = 1, 2, ..., 200:")
for n, a_n in zip(n_full, a_n_full):
    print(f"a_{n} = {a_n:.6f}")
```

#$c).$#

```python
plt.figure(figsize=(10, 6))
plt.plot(n_full, a_n_full, label='$a_n = n \\sin(1/n)$')
plt.axhline(y=1, color='r', linestyle='--', label='$y = 1$')
plt.xlabel('n')
plt.ylabel('$a_n$')
plt.title('Gráfico de $a_n$ y la línea horizontal $y = 1$')
plt.legend()
plt.grid(True)
plt.show()
```

#(d) Analizar el límite de $a_n$ a medida que $n$ tiende a infinito.#
##Observamos que $a_n$ se aproxima a 1 en el gráfico, pero para una confirmación más precisa,podemos verificar con un gran valor de $n$. ##

#$e).$#

```python
n_greater_than_999 = n_full[a_n_full > 0.999][0]
print(f"El valor más pequeño de n para el cual a_n > 0.999 es: {n_greater_than_999}")
```

#$f).$#

```python
n_within_tolerance = n_full[np.abs(a_n_full - 1) < 0.0001][0]
print(f"El primer valor de n para el cual |a_n - 1| < 0.0001 es: {n_within_tolerance}")
```
