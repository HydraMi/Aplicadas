#Ejercicio 3#

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 100, 1)
sin_n_div_n = np.sin(n) / n

plt.plot(n, sin_n_div_n, label=r'$\frac{\sin(n)}{n}$')
plt.hlines(0, 1, 100, colors='r', label=r'$y = 0$', linestyles='dashed')
plt.xlabel('n')
plt.ylabel(r'$\frac{\sin(n)}{n}$')
plt.title(r'Comportamiento de $\frac{\sin(n)}{n}$ para grandes valores de $n$')
plt.legend()
plt.show()
```

```python
import numpy as np

n_values = np.arange(1, 201)

a_n_values = [(1 + 1/n)**n for n in n_values]

for n, a_n in zip(n_values, a_n_values):
    if a_n > 2.7:
        print(f"El menor valor de n tal que a_n > 2.7 es: {n}")
        break
```

##f).##

```python
import numpy as np

e = np.exp(1)

n_values = np.arange(1, 201)

a_n_values = [(1 + 1/n)**n for n in n_values]

for n, a_n in zip(n_values, a_n_values):
    if abs(a_n - e) < 0.01:
        print(f"El primer valor de n tal que |a_n - e| < 0.01 es: {n}")
        break
```
